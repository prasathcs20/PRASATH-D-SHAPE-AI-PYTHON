```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data= pd.read_csv("gas_prices.csv")
data
```

```python
data.info()
```

```python
data.isnull()
```

```python
data.describe()
```

```python
x= data["Year"]
y= data["France"]
```

```python
plt.xlabel("YEAR")
plt.ylabel("GAS PRICE")
plt.title("Bar graph ")
plt.bar(x,y,color="green",width=0.5)
```

```python
plt.title("scatterplot")
plt.xlabel("YEARS")
plt.ylabel("France")

plt.scatter(x,y,color="green",marker="*",label="FRANCE")
plt.scatter(x,data["USA"],color="red",label="USA")
plt.legend()
plt.show()
```

```python
countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
for country in data:
    if country in countries_to_look_at:
         plt.plot(data.Year, data[country], marker='.')
plt.xlabel('Year')
plt.ylabel('PRICE')
plt.title("line chart")
plt.show()
```

```python
corr = data.corr()
sns.heatmap(corr)
```

```python
n=data[data['South Korea']>4]
plt.hist(n,x)
plt.title('HISTOGRAM')
plt.show()
```
